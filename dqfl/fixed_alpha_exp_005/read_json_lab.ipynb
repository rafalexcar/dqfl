```python
import os
import re

log_file = "experiment_log_0.json/experiment_log.json"
# Check if the log file exists
if not os.path.exists(log_file):
    print(f"{log_file} does not exist.")

# Read the logs and extract accuracies
with open(log_file, "r") as file:
    log_text = file.read()
```

```python
print(log_text)
```

```python

# Regular expression to match accuracy values
pattern = r"'avr_accuracy': \[(.*?)\]"

# Search for matches after '[SUMMARY]'
match = re.findall(pattern, log_text, re.DOTALL)

print(match[0])
```

```python
import ast

runs = {}
for run in range(len(match)):
    extracted_string = match[run].strip().replace('\n', '').replace('\t', '')
    avr_accuracy_list = ast.literal_eval(f"[{extracted_string}]")
    runs[run] = avr_accuracy_list
```

```python
runs
```

```python
import numpy as np
# Transform to a NumPy array
# Extract the second values only
array = np.array([
    [value[1] for value in values]
    for values in runs.values()
])

print(array)
```
